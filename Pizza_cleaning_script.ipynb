## Grouping 

```python
#Group by [pizza size id]and get the count of sales(row count)
grouped_df_pizza_size=pizza_sales_df.groupby(['pizza_size_id']).count()
grouped_df_pizza_size
```

```python
#Group by [pizza size id]and get the sum of sales(row count)
grouped_df_pizza_size_by_sum=pizza_sales_df.groupby(['pizza_size_id'])['total_price'].sum()
grouped_df_pizza_size_by_sum
```

```python
#Group by [pizza size id]and get the sum total price and coding
grouped_df_pizza_size_sales_quantity=pizza_sales_df.groupby(['pizza_size_id'])[['total_price','quantity']].sum()
grouped_df_pizza_size_sales_quantity
```

```python
#Lokking at diffierent aggregate function

#count the no of n/ null value for each group

grouped_df_count=pizza_sales_df.groupby(['pizza_size_id'])[['total_price','quantity']].count()
grouped_df_count
```

```python
#sum the values in each group
grouped_df_sum=pizza_sales_df.groupby(['pizza_size_id'])[['total_price','quantity']].sum()
grouped_df_sum
```

```python
#average the values in each group
grouped_df_agg=pizza_sales_df.groupby(['pizza_size_id'])[['total_price','quantity']].mean()
grouped_df_agg
```

```python
#using agg to perform diffierent aggregations on diffierent columns
aggregated_data=pizza_sales_df.groupby(['pizza_size_id']).agg({'quantity':'sum','total_price':'mean'})
aggregated_data
```
